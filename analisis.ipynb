```python
import pandas as pd

datos_raw = pd.read_excel('datos pendulo multiple csv.xlsx', na_values='N/A')
datos_raw.head()
```

```python
# se crean 3 dataframes, uno para angulo diferentes

angulo_10 = datos_raw[(datos_raw['Angulo']==10)]
angulo_30 = datos_raw[(datos_raw['Angulo']==30)]
angulo_45 = datos_raw[(datos_raw['Angulo']==45)]
```

con los datos anteriores se hacen las siguientes graficas de caja usando matplotlib

### Para 10°

```python
angulo_10['Pendulo simple (altura)'].min()
```

```python
import numpy as np
import seaborn as sns
import scipy
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
grafico_10 = sns.regplot(data=angulo_10,x='Tiempo para 1 oscilación (s)', y='Pendulo simple (altura)')

#regresion y su ecuacion con scipy
slope, intercept, r, p, sterr = scipy.stats.linregress(x=grafico_10.get_lines()[0].get_xdata(),
                                                      y=grafico_10.get_lines()[0].get_ydata())

grafico_10.set_xlim(0.81,1.1)
grafico_10.set_ylim(0.85,1.21)
plt.title('Periodo experimental vs Periodo teorico - Angulo 10°')
plt.text(s=f'y = {round(slope,3)}x + {round(intercept,3)}\nR^2={r**2}\nvalor P={p}',x=0.2,y=0.8, fontsize=9, horizontalalignment='center',verticalalignment='center',transform = grafico_10.transAxes)
```

```python
grafico_45 = sns.regplot(data=angulo_45,x='Tiempo para 1 oscilación (s)', y='Pendulo simple (altura)')

#regresion y su ecuacion con scipy
slope, intercept, r, p, sterr = scipy.stats.linregress(x=grafico_45.get_lines()[0].get_xdata(),
                                                      y=grafico_45.get_lines()[0].get_ydata())

grafico_45.set_xlim(0.89,1.21)
grafico_45.set_ylim(0.88,1.2)
plt.title('Periodo experimental vs Periodo teorico - Angulo 30°')
plt.text(s=f'y = {round(slope,3)}x + {round(intercept,3)}\nR^2={r**2}\nvalor P={p}',x=0.2,y=0.8, fontsize=9, horizontalalignment='center',verticalalignment='center',transform = grafico_45.transAxes)
```

Viendo el coeficiente R^2 casi que equivalente a 1, además que las 3 ecuaciones tienen una pendiente menor a 1.2, puese inferirse que la relación entre el periodo obtenido experimentalmente y el periodo teórico teniendo en cuenta la altura del péndulo es casi perfectamente lineal. Es decir, parece ser que los péndulos con multiples cuerdas se comportan igual que el correspondiente péndulo simple cuya longitud de la cuerda es igual a la altura desde el vértice (donde se halla el péndulo) perpendicularmente hacia arriba
