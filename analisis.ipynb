```python
import pandas as pd

datos_raw = pd.read_excel('datos pendulo multiple csv.xlsx', na_values='N/A')
datos_raw.head()
```

```python
# se crean 3 dataframes, uno para angulo diferentes

angulo_10 = datos_raw[(datos_raw['Angulo']==10)]
angulo_30 = datos_raw[(datos_raw['Angulo']==30)]
angulo_45 = datos_raw[(datos_raw['Angulo']==45)]
```

con los datos anteriores se hacen las siguientes graficas de caja usando matplotlib

```python
import numpy as np
import seaborn as sns
import scipy
import matplotlib.pyplot as plt
```

```python
grafico_10 = sns.regplot(data=angulo_10,x='Tiempo para 1 oscilación (s)', y='Pendulo simple (altura)')

#regresion y su ecuacion con scipy
slope, intercept, r, p, sterr = scipy.stats.linregress(x=grafico_10.get_lines()[0].get_xdata(),
                                                      y=grafico_10.get_lines()[0].get_ydata())

grafico_10.set_xlim(0.81,1.1)
grafico_10.set_ylim(0.85,1.21)
plt.title('Periodo experimental vs Periodo teórico - Ángulo 10°')
plt.text(s=f'y = {round(slope,3)}x + {round(intercept,3)}\nR^2={0.9017}',x=0.2,y=0.8, fontsize=9, horizontalalignment='center',verticalalignment='center',transform = grafico_10.transAxes)
plt.xlabel("Periodo Experimental (s)")
plt.ylabel("Periodo Teórico (s)")
```

```python
angulo_30['Pendulo simple (altura)'].min()
```

```python
grafico_30 = sns.regplot(data=angulo_30,x='Tiempo para 1 oscilación (s)', y='Pendulo simple (altura)')

#regresion y su ecuacion con scipy
slope, intercept, r, p, sterr = scipy.stats.linregress(x=grafico_30.get_lines()[0].get_xdata(),
                                                      y=grafico_30.get_lines()[0].get_ydata())

grafico_30.set_xlim(0.88,1.21)
grafico_30.set_ylim(0.88,1.2)
plt.title('Periodo experimental vs Periodo teórico - Ángulo 30°')
plt.text(s=f'y = {round(slope,3)}x + {round(intercept,3)}\nR^2={0.9556}',x=0.2,y=0.8, fontsize=9, horizontalalignment='center',verticalalignment='center',transform = grafico_30.transAxes)
plt.xlabel("Periodo Experimental (s)")
plt.ylabel("Periodo Teórico (s)")
```

```python
angulo_45['Pendulo simple (altura)'].min()
```

```python
grafico_45 = sns.regplot(data=angulo_45,x='Tiempo para 1 oscilación (s)', y='Pendulo simple (altura)')

#regresion y su ecuacion con scipy
slope, intercept, r, p, sterr = scipy.stats.linregress(x=grafico_45.get_lines()[0].get_xdata(),
                                                      y=grafico_45.get_lines()[0].get_ydata())

grafico_45.set_xlim(0.89,1.21)
grafico_45.set_ylim(0.88,1.2)
plt.title('Periodo experimental vs Periodo teórico - Ángulo 45°')
plt.text(s=f'y = {round(slope,3)}x + {round(intercept,3)}\nR^2={0.9777}',x=0.2,y=0.8, fontsize=9, horizontalalignment='center',verticalalignment='center',transform = grafico_45.transAxes)
plt.xlabel("Periodo Experimental (s)")
plt.ylabel("Periodo Teórico (s)")
```

Viendo el coeficiente R^2 casi que equivalente a 1, además que las 3 ecuaciones tienen una pendiente menor a 1.2, puese inferirse que la relación entre el periodo obtenido experimentalmente y el periodo teórico teniendo en cuenta la altura del péndulo es casi perfectamente lineal. Es decir, parece ser que los péndulos con multiples cuerdas se comportan igual que el correspondiente péndulo simple cuya longitud de la cuerda es igual a la altura desde el vértice (donde se halla el péndulo) perpendicularmente hacia arriba

Así, el promedio de oscilación para los 10, 30 y 45 grados es el siguiente

```python
promedio_10 = angulo_10['Tiempo para 1 oscilación (s)'].mean()
promedio_30 = angulo_30['Tiempo para 1 oscilación (s)'].mean()
promedio_45 = angulo_45['Tiempo para 1 oscilación (s)'].mean()

print(f'Promedio experimental para 10 grados={promedio_10}\nPromedio experimental para 30 grados={promedio_30}\nPromedio experimental para 45 grados={promedio_45}')
```

```python
teorico_10 = angulo_10['Pendulo simple (altura)'].mean()
teorico_30 = angulo_30['Pendulo simple (altura)'].mean()
teorico_45 = angulo_45['Pendulo simple (altura)'].mean()

print(f'Promedio teorico para 10 grados= {teorico_10}\nPromedio teorico para 30 grados= {teorico_30}\nPromedio teorico para 45 grados= {teorico_45 }')
```

error relativo entre medida experimetal y promedio del periodo es el siguiente

```python
err_relativo_30 = round((abs(teorico_10-promedio_10)/teorico_30)*100,2)
err_relativo_10 = round((abs(teorico_30-promedio_30)/teorico_10)*100,2)
err_relativo_45 = round((abs(teorico_45-promedio_45)/teorico_45)*100,2)

print(f'error relativo para 10 grados= {err_relativo_10}%\nerror relativo para 30 grados= {err_relativo_30}%\nerror relativo para 45 grados= {err_relativo_45 }%')

print(f'\nError relativo promedio {round((err_relativo_10+err_relativo_30+err_relativo_45)/3,2)}%')
```

un error relatrivo de 4.11% indica que hay una alta relación entre el periodo del péndulo simple y la altura de un péndulo con múltiples cuerdas
